```python
#importing the libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as mtp

!pip install apyori


#importing the dataset
dataset=pd.read_csv('/kaggle/input/market-basket-optimization/Market_Basket_Optimisation.csv')
transactions=[]
for i in range(0, 7500):  # Instead of 7501
    transactions.append([str(dataset.values[i, j]) for j in range(0, 20)])

#traning the dataset on the apriori model
from apyori import apriori
rules=apriori(transactions=transactions,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2,max_length=2)
#displaying the result
results=list(rules)
results

#visualising the rule,support,confidence and lift
for item in results:
    pair = item[0]
    items=[x for x in pair]
    print("RULE:"+ items[0] +"->"+ items[1])

    print("SUPPORT:"+str(item[1]))
    print("CONFIDENCE:"+str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))  
    print("=====================================") 
```
